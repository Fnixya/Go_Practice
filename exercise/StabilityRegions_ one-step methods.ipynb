```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt

x = [0, 0]
y = [-8, 8]
K = 'k'
plt.plot(y, x, K)
plt.plot(x, y, K)

C = 'color'
c = ['b', 'r', 'g', 'm', 'c', 'y']
x = np.arange(-6, 6, 0.01)
y = np.arange(-6, 6, 0.01)
x, y = np.meshgrid(x, y)


z = x + 1j * y
R1 = 1 + z
R2 = 1 + z + 0.5 * z**2
R3 = 1 + z + 0.5 * z**2 + (1/6) * z**3
R4 = 1 + z + 0.5 * z**2 + (1/6) * z**3 + (1/24) * z**4
R10 = 1 + z + 0.5 * z**2 + (1/6) * z**3 + (1/24) * z**4 + (1/120) * z**5 + (1/720) * z**6 + (1/5040) * z**7 + (1/40320) * z**8 + (1/362880) * z**9 + (1/3628800) * z**10 # + (1/39916800) * z**11 + (1/479001600) * z**12 + (1/6227020800) * z**13 + (1/87178291200) * z**14 + (1/1307674368000) * z**15 + (1/20922789888000) * z**16

A = 1
C9 = (-797/50803200 + (1/25200)*A + (37/4233600)*sqrt(21)*A - (499/152409600)*sqrt(21)) * z**9
C10 = (1/470400 + (1/2083725)*sqrt(21) - (31/940800)*A - (61/8467200)*sqrt(21)*A) * z**10
C11 = (-1/29030400 - (13/4267468800)*sqrt(21) + (11/1612800)*A + (353/237081600 )*sqrt(21)*A) * z**11
R8 = 1 + z + 0.5 * z**2 + (1/6) * z**3 + (1/24) * z**4 + (1/120) * z**5 + (1/720) * z**6 + (1/5040) * z**7 + (1/40320) * z**8 + C9 + C10 + C11

plt.contour(x, y, abs(R1), levels=[1], colors=c[0], linewidths=1)
plt.contour(x, y, abs(R2), levels=[1], colors=c[1], linewidths=1)
plt.contour(x, y, abs(R3), levels=[1], colors=c[2], linewidths=1)
plt.contour(x, y, abs(R4), levels=[1], colors=c[3], linewidths=1)
plt.contour(x, y, abs(R8), levels=[1], colors=c[4], linewidths=2)
plt.contour(x, y, abs(R10), levels=[1], colors=c[5], linewidths=2)

plt.axis('square')
plt.grid(True)
plt.title('Runge-Kutta orders 1, 2, 3, 4, 8', fontsize=16)
plt.axis([-5, 2, -3.5, 3.5])

plt.show()
```
